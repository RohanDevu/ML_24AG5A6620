```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.array([4, 8])
y = np.array([3, 10])
plt.plot(x,y)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,4,9,15,18]
plt.figure(figsize=(8,5))
plt.plot(x,y,color='brown',linestyle='--',marker='o',linewidth=2)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("line chart")
plt.legend(['my data series'])
plt.show()
```

```python
categories=['A','B','C','D','E']
values=[5,4,7,13,10]
plt.figure(figsize=(8,5))
plt.bar(categories,values,color='brown')
plt.xlabel('categories')
plt.ylabel('values')
plt.title('Bar Chart')
plt.show()
```

```python
#sample data
data=[15,11,10,17,19,25,24,12,16,14,18,11]
#create histogram
plt.hist(data,bins=12,edgecolor='black')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Histogram')
plt.show()
```

```python
#sample data
x=[1,2,3,4,5]
y=[5,7,4,8,6]
#create scatter plot
plt.scatter(x,y)
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Simple Scatter Plot Example')
plt.show()
```

```python
#sample data
labels=['a','b','c','d','e']
sizes=[15,32,18,17,18]
colors=['white','lightgreen','orange','pink','skyblue']
#create pie chart
plt.pie(sizes,colors=colors, labels=labels,autopct='%1.1f%%',startangle=140)
plt.title('Simple Pie Chart Example')
plt.axis('equal')
plt.show()
```

```python
#sample data:10x10 matrix
data =np.random.rand(10,10)
#create heatmap
plt.imshow(data,cmap='viridis',interpolation='nearest')
plt.colorbar() #show color scale
plt.title('Simple Heatmap Example')
plt.show()
```
